```python
# First I'll calculate the measure of roughness (mr) for a column of characters (letters in this case)
# summation of letters 1-26 fi*(fi-1)/L*(L-1)
def calc_mr(character_frequency, all_characters):
    numerator = sum(f * (f - 1) for f in character_frequency)
    denominator = all_characters * (all_characters - 1)

    if denominator == 0:
        return  0
    
    return numerator/denominator

#Creating a matrix for each guessed key length
def keylength_matrix(cipher, gkey_len):
    
    #creating the columns for the matrix based on the guessed keylength
    columns = ['' for _ in range(gkey_len)]

    #distributing characters into columns of the matrix
    for i, char in enumerate(cipher):
        columns[i % gkey_len] += char

    #output the columns for a guessed key length
    print(f"\nKey Length: {gkey_len}")
    print("Cipher Matrix:")
    for idx, col in enumerate(columns):
        print(f"Column {idx + 1}: {col}")

    return columns

# Calculating the average measure of roughness for a guessed key_length
def mr_for_gkeylen(cipher, gkey_len):
    columns = keylength_matrix(cipher, gkey_len)
    mr_values = []

    for column in columns:
        # how frequent does each letter appear in a column?
        frequencies = {chr(i): 0 for i in range(65, 91)} #ASCII values for A and Z
        for char in column:
            if char in frequencies:
                frequencies[char] += 1

        # list of letters that appear frequently and their total number
        character_frequency = [frequencies[chr(i)] for i in range(65, 91)]
        all_characters = sum(character_frequency)

        # calculating thr MR for a particular column
        mr = calc_mr(character_frequency, all_characters)
        mr_values.append(mr)


    # calculating the average measure of roughness for a guessed key length
    average_mr = sum(mr_values) / len(mr_values) if mr_values else 0
    print(f"The average MR for the guessed keylength {gkey_len}: {average_mr}")
    return average_mr


#Predicting the key length for the cipher (the averahe mr closest to the mr for english=0.0686)
def pred_keylength(cipher, guessed_keylength, E_mr=0.0686):
    results = {}

    for gkey_len in guessed_keylength:
        average_mr = mr_for_gkeylen(cipher, gkey_len)
        results[gkey_len] = average_mr

    # Finding the key length with the average MR nearest to the English MR
    nearest_gkeylength = min(results, key=lambda k: abs(results[k] - E_mr))

    print(f"\nThe Predicted Key Length is:{nearest_gkeylength}")
    return nearest_gkeylength



#Trying it out
cipher = "MxhuklooezspoigpvslipyoxkhsbzowdvvjhobsxswhukayfyjfiabardovtwdoyribhugrdvrkcoqbapngrkpkmgnikbpoixhiopvotkspSikvgcbjRyanvifsejizwpzinwgypyqnzmybnthxcgxefsykvojrxvohhxrornlxofqowmcikvsbtoxDvrbmvznmibgblIfsecsyryozorvtlkfzurihrthsbtzloweivydfmedvrxmxusxswhukjyfryxkbqlmcvvtkpfbsxrsfopfsexmfseyxrogiydhuxseuuzlojnrpomOaxgvnzqkrrKzofjusnheapiskzvkcejmxoeeakgazmdgvysvogosxwgcecwgyqegviQegviakggnivwskfvcbjspSikvgcbjIfseeqyfaorqhukzszygkoffcsusguxrsfujdggxesbfujpzhzicqnxvssqurdvrcmxrnthkhqawuhukwyiajspggxmxufcsezqryvzgniwhbyposcZloweorcheaqobgyaofrtxwsekaycqgrnarzevhukcgsekevwikkbcjtjbczzlosailkbgkhdfrkwdvnzpsbrjxrssuvoggyXrsfzvsbtysphukmbzhzickrxigcikrpfbszsbryxrogisezqnywoajxrssrydsfiebjrjjbczzlognivorRbibkbuhdfrkwgchrhybyetvolojdvrsycwpoexgukebhjgwziekXrsfurqgblIfsecsyrjkvobgpychrtxofggmxartxdvreaofrcsfsaordcgnipooxmmcsrmpsvzwozsZlofrcecofurqhbhvsbtzlofnorkarrsnmgulooycsebqyexrrbixofujdzhrpkplzsccbzlohukaszqgrsanrwytgnipcekwdPhzxrszuwddbcibthrspoyrakggniCcamspSgkvxwgeewsyuhihugxrsyjxrsogpkbpkfohjkixhukzszygkooajxrssuvoggOxgofyesrgnedwszloGbtkytRzibbvzcgofkzofoxsusazlovnxqyblhidkrkrwoagrnbnzybsjuyvrfnedhrxexrRbibkbuhgchrhpoyrmxhbxysbSuvmsazybwryxrsiopvotkvcvnjpsjrjmxdrggohukmbahymmbrbibtnrxofvtkdvrYsxublIdsetmdmnrakmfnikfqArdwyurornearsayswsgnmxupnexurjPowygakgfobdsrtarsaylotvxwdvrgvnhukwszrtgoGukwdcbjedhukinurujdvrKzofjusntbxichukvroajwlfhylsbtzlopnxoytgnicopxinhekiczvyxobvtkdcgnigwajFehfuqohuorqknyabcamJyfnypybtgwcvrisezqxiwszhibhukxbsrylkruaqwsqcmdvzawsqnrsgurtxvsuaqvwxkewcgnibgukebhokedoyceigcxicsazevknewmczlsbhvtkLigtsgofylodekwcsqnibsnxxyhukfkfxzlofrcecbbzlsbtRisznlvykakhkbqzybbrjxyknxhdvrbmvznmiCvrnenprkrlcetashugkstggvkfrzevsazjyfukebwamxrszawsqblxrsjuvvrzuvoyrkrvmgnexoaesxsrrwoKuopocgnibgukebrzkpyrvkwVsvrersnxhdvrtsdsfhixsnzldvrclsgckvccszlosnxxrhuklooezfoogujdvrzvosfzlotyaxdseujlwejasbtyExrauagvrxidvrxicvbapnvnbilsrteccamxrsekakgbtpioaarcsgzpsbtwyssgYlovhxvssqhemyguxrsiopvotklofskidgxoqwwamxrsfujdtbxichsrsyfNywrsnvtbcnilorgnifwyreqsfwykfrylognceqfbatqognibsqgvyiajQkggkvOzlurdvrbmvznmicabyxbsikvorzawsqvgrkbqqiodrxsphukWybtujOhrxrshlNmctniigofvevsuowroajwdfrsfvwamechukcmzhzgrsqzloggxmxufujrwfrydsGniccamIvmbtwfcviidfrsfvsqhebsyeelcikegvvytofVzwqcakEwiesybfvvtvsqzlbchmldvrivykqRisznvycvrjlofjgcpcecebrukvrsnxxzchthsbtMsxsUuamchrhshokkybrSechrxIvmbtarogjsichsikbfnikgxkhrsebssqrzmqvgcmdvskebSyesxzbuoorhvedvrxlsgreickvjigwgnwyfeuaDvrYsxublIdsetmdmViexbbrsxurxlooeoxSheoinhbvpkmvzxrwfssbbvtkkgVnefsqurosikvirnespalrmpsoaxxcgnmxupgqoHukwdfvtkcoekwszrtxVsvrecpekedvpgyqvgorrsezlbcnzXrsFurqcsKxofaoxiknyxrsikvisfyixqrujOjrxaycqCmdvbaxshgniloygrmsokxgsrtxrsiopvotkexrgnipcekwdkbapnqeaqlzrZloqeutckbapntnopdvrxmfseyayiyjhbmnthdvrkrmvntxorgxiogjuyvrqoiDvrxiwifzfogbsidvvtkgspgrncYkmvofgmnvrxzywpkvsgvtkgwgnyburtgiAnefoanefohukwybtowrwqjixcehvyyrtAoqntjslvzvsuuzIvmbtwrcbqlsgukenWikxbwrjMfscreisqkzofltsdsrbibmpnsbroaxdvrsivcqelkgfrmzdrjegolGrnkvzlyigzlogbtkOjrxaycqcmvzsghoHukzszygkoffrsyyrjedcakexcgnibhukmbtniicdnrigwgnjooeKzofjusnvnjrojrxoxcjthsgnyxofGniivnjevknewlsrttbcgkgdsqgpgolywktrHydbbcxrsikvitbarnogosxcszlowermfsfcecgyotzwamxrfbakrhukmbtvtkoffRisznxipifkhdcnigodgoxDvrxiroqzslsnceiGukgyiyjwdwyrlooezlofrsrkbgysphukwybtymxhukesfsgmxhnyxrslcibsNthstgnibsjgwojrtegvvytofblqegvipotgzlofrcecvbviDvnzrsuuzarwykxrsiopvotkwvsczPowygqkrrghoqvymybFnigchrhqcguxrsukebhblxrssuvoggzlognivortrensjnibsgnipweyxccamakgouvxHukivrrxwcovjmdknyezzniiyttxikhcuaofjnibsgniwifogytgniooezlpzbcinzvqikfvbibWszloGbtkytRzibbvzcmchrhlssuyxrntcgvrxishjuyvrokxrsekAshutsdvvtklignibgzgpvtyaxoqnxzorsxswhukaycqujkbRbibkbuhdfrkwrsfrmzdrjmxhbzlotbxichhthofgnimcikvytqgvubrywDvrssybyokrhsopdsekhdveuyqvgnilfntgrsfiechvtkcwybibgughykfurdvrmvyiajPowygwrsnxxzchthorvtlofpnichauxpfbsjooehydteuqdvrcisuuzspkugxcvrcecoouydhbjsKgfnigoyqinrrktofvtxyhukjyfryxdvrymvsaiiqfrclooioibHukycinrgrceawytpxmmyrzwkbquavgjgwkpfkrdoajifsazlokvthcsrsinhbnsvrvzwlfrgxrZropktbrpykrjxrscgxrguklkrxtsgbformspnmvruusnphzxybvmldwglivhhtjkavrmkfyooohukjyfryxshfkppvnjwrwszinVbavcdnywornthchvrpcvrcevyrjyxhvrjsbnrpigukvoopninhukwkqekhqznjiShjgwkdyggocshvoognxkyvtklsnaxihukxbsryxkzykvkbqupnsezlkbntcgvrxiozfkmxhukjyfryxdvrovlfntgrsfordsezasbrjpsyrzlotvtkoffujqcqyMxhukgobgkvytgniqznjichbuhdvrZvosblWybtymdgogvuuyuasbtlesbgrcsbgniwcbtpsuuzPowygezdeuemvrjxrsgxioqnaxschypivrxlooezmxvrxxrfbgxCvrvvogfkhrsenexrnmesbfzxrsogvuoajgvcfkhrsekcogyowdsaorqtbxxrsuaqytgniwifogcvrnenoyceigxtsgbOaxdvrxigoftsdvvtktifzexszvxiopnmxufopobpkXooeytbwpqinvrxiisfNsgqbapnhukwybthiqcakWrsughlsyoifsqcmdvrbibmsofofbllofokmxugnedgukayiyjjsbqoxrsekFehauachnthsbthipcekxrsfggbsqzvosfnipsyzqyfrrschgnexsikvSbukvnsfvesffnicoaqxyhukkbchthkbqvyvzrjsehukvpzhziShjgwkzyylovnjpotgYlofnoworvzxyvrxpsdfgrnprmexhbvpkmcuybwamevzukvccexsgvrxjooegrnvrxpyjrlsbSikvgcbjmxhbzloarrsnmGnixcgkwgsekwytggrnhrtxkhvbikhsovchoaxkgfnizzneinhukcqfrcwdfbtkofzuvoqrxxkwaGrnhukrcczkxrwamihhegsbrvtebmugtzsakhDvrgmboeuyxrukvlstgrdcuaqxcgcmdvukvwifogligcmdvnjiodrxsvrrxqozbjcShjgwpovtxvwxkxrsrilycsghsgggrdgbtkligoxgofzlofrRisznylooezpooczWrscreisqlechrxxrsauxogfuebwammxhbzlobvmldoajxrsuaqqfrcpyiqkvpwyrmxugniqznjigwgnecchthdvnzakgouxrtnsmvwnxexrakaDvrzvosfyakmrjxrsvxfboailogjnmcdrxmxuvtxsarcmdvgniwifogDvrmvyiajfobrgxrvrxjosgzvoaorinoajxrsnovshfkppgrkqorguwrwzsibZropkgsrydsjgwxcyurqsezlogbavmsblxrszawsqvzakgnisxrhoxmvntrozvtkdvrysxugnedvnjfosanmnrrthosccmdvvtxrsrgvdvGniCcamspSgkvxwgeakgazpyggOxroqhiobjgmdwamakwgorqtbxwyarurokuugyiyjlooeoxgvbisezqhvsbtoxlopqxyzvliKggnipwagpxcgklebtordvrgmbZropkzbcibsqnibtyaxovrxgrsfzlooiorqkvzldvrkjpcezXrstrensjgwkzvbigwgnqegvirykgnidfrkwcwammxugnigwajgkfeemxugniwsyuhitnxexrjohoOajmxhugxwczkrdgukoxsjzlopnrexqrnenprkrbsfzsbsqKzofjusnknywktr ".upper()
guessed_keylength = [4, 5, 6]
predicted_keylength = pred_keylength(cipher, guessed_keylength)
```
